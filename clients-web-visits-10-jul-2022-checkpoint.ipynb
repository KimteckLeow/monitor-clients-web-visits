```python
# Import necessary packages
from matplotlib.axes._axes import _log as matplotlib_axes_logger
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import earthpy as et
import numpy as np

# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Dealing with error thrown by one of the plots
matplotlib_axes_logger.setLevel('ERROR')
import warnings
warnings.filterwarnings('ignore')

# Adjust font size and style of all plots in notebook with seaborn
sns.set(font_scale=1.5, style="whitegrid")

# Set working directory to earth-analytics
os.chdir(os.path.join(et.io.HOME, 
                      "earth-analytics"))

# monitor-clients-web-visits/clients-visits-web.csv
# Import data from .csv file 
fname = os.path.join("monitor-clients-web-visits",
                     "clients-visits-web.csv")

# Import data using datetime and set index to datetime
# clients_visits_web = pd.read_csv(fname, parse_dates=['date'],index_col=['date'])

clients_visits_web = pd.read_csv(fname)
clients_visits_web
```

```python
# Cleaning data-getting rid of rows where no UserId is shown

# R: ClientVisits[ ClientVisits$ClientId != "0", ]

clients_visits_with_id = clients_visits_web[clients_visits_web["UserId"] != "0"]
clients_visits_with_id
```

```python
# ClientVisits[ClientVisits$Date >= as.Date("2015-01-01"), ]

clients_visits_with_id_2015 = clients_visits_with_id[clients_visits_with_id["Date"] >= "2015-01-01"]
clients_visits_with_id_2015
```

```python
# user_df[user_df['sign_up_date'].str.contains('2022')]

clients_visits_pages = clients_visits_with_id_2015[clients_visits_with_id_2015["Page"] .str.contains('/client/product/')]
clients_visits_pages
```

```python
# print(df['a'].str[:2])
# The product's code is the set of ten characters right after the "/client/product/" string
# The numbers of 17-26 were obtained by merely counting the characters. 
# The product code starts on the 17th and finishes on the 26th character.

print(clients_visits_pages['Page'].str[17 : 27]) 

             
```

```python
type(clients_visits_pages['Page'].str[7 : 27])
```

```python
productid = clients_visits_pages['Page'].str[7 : 27].to_frame()
productid
```

```python
# counting the repeating product id
productid_occurence = productid.pivot_table(index = ['Page'], aggfunc ='size')
productid_occurence    
    
# displaying the duplicate Series
# print(productid_occurence)
```

```python
type(productid_occurence)
```

```python
productid_occurence_df = productid_occurence.to_frame() 
productid_occurence_df
```

```python
productid_occurence_df.reset_index(inplace=True)
productid_occurence_df
```

```python
productid_occurence_top10 = productid_occurence_df_sort.head(10)
productid_occurence_top10
```

```python
productid_occurence_top10.reset_index(inplace=True)
productid_occurence_top10
```

```python
productid_occurence_top10.info()
```

```python
# Import pyplot
import matplotlib.pyplot as plt

      
# Define plot space
fig, ax = plt.subplots(figsize=(30, 15))

# Define x and y axes
# plt.yscale("log")

plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

ax.bar(productid_occurence_top10['page_visit'], 
       productid_occurence_top10['occurence'],
       color = 'blue', 
       edgecolor = 'darkblue')

# Set plot title and axes labels
ax.set(title = "Top 10 of product visits in 2015",
       xlabel = "Product code",
       ylabel = "Number of visits")

ax.set(ylim=[0, 200000])

plt.title('Top 10 of product visits in 2015', fontdict={'fontsize':50})
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
plt.xticks(rotation = 45)

plt.show()
```
